### 1. ¿Cómo se distribuye la cantidad de vuelos reservados por mes durante el año?

El Line Plot es el gráfico correcto porque conecta los puntos en orden cronológico (Mes 1 al Mes 12) y muestra la tendencia de la serie temporal.

El gráfico de líneas revela una clara tendencia estacional en la reserva de vuelos:

1. Pico de Temporada Alta (Julio y Agosto): Los meses de Julio y Agosto muestran el punto más alto en el total de vuelos, lo cual es típico de la temporada de vacaciones de verano.

2. Temporada Baja (Enero/Febrero): Los meses de Enero y Febrero registran la menor cantidad de vuelos, marcando la temporada baja después de las vacaciones de fin de año.

3. Crecimiento en Primavera: Hay un aumento constante y pronunciado en la reserva de vuelos a partir de Marzo hasta el pico de verano, indicando la temporada de primavera/verano.

Conclusión: Existe una fuerte estacionalidad donde el volumen de vuelos está directamente relacionado con los periodos vacacionales.

```python
plt.figure(figsize=(12, 6))

(
    df_customer_clean.groupby('Month')['Flights Booked'] 
    .sum()                                      
    .plot(kind='line', marker='o', color=pal.ocean_blue_serenity["Azul_Profundo"]) 
)

plt.title('Tendencia de Vuelos Reservados por Mes')
plt.ylabel('Total de Vuelos')
plt.xlabel('Mes')
plt.show()
```

### 2. ¿Existe una relación entre la distancia de los vuelos y los puntos acumulados por los cliente?


El Scatter plot es el gráfico correcto porque son dos variables de las que se necesita ver la dispersión. Poca dispersión indica que ambas variables están altamente vinculadas.
Cada punto es un cliente. 


1. Existen líneas diagonales fuertes. Correlación positiva con poca dispersión entre las dos variables.
2. A mayor distancia volada, mayor es la acumulación de puntos (la acumulación de puntos depende directamente de las millas voladas).
3. Existen varias líneas porque el sistema de acumulación de puntos es diferente para cada tipo de cliente ("Loyalty Card" y "Enrollment Type").
4. Fuerte relación entre el tipo de "Loyalty Card" y la acumulación de puntos. 
5. El "Enrollment Type" "2018 promotion" acumula más puntos (líneas superiores). 

```python

# 1. Crear la figura y los dos subplots (1 fila, 2 columnas)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 7)) 


#Scatter Plot 1: Segmentado por Tarjeta de Lealtad (Loyalty Card)
sns.scatterplot(
    data=df_customer_clean,
    x='Distance',        
    y='Points Accumulated', 
    hue='Loyalty Card',   
    alpha=0.6,           
    palette='Set1',
    ax=axes[0]             # Asignar al primer subplot (índice 0)
)
axes[0].set_title('Distancia vs. Puntos (por Tarjeta de Lealtad)')
axes[0].set_xlabel('Distancia Total Volada')
axes[0].set_ylabel('Puntos Acumulados')
axes[0].legend(title='Tarjeta de Lealtad')
axes[0].grid(True, linestyle='--', alpha=0.5)


# B. Scatter Plot 2: Segmentado por Tipo de Inscripción (Enrollment Type)

sns.scatterplot(
    data=df_customer_clean,
    x='Distance',        
    y='Points Accumulated', 
    hue='Enrollment Type', 
    alpha=0.6,           
    palette='Dark2',
    ax=axes[1]             # Asignar al segundo subplot (índice 1)
)
axes[1].set_title('Distancia vs. Puntos (por Tipo de Inscripción)')
axes[1].set_xlabel('Distancia Total Volada')
axes[1].set_ylabel('Puntos Acumulados')
axes[1].legend(title='Tipo de Inscripción')
axes[1].grid(True, linestyle='--', alpha=0.5)

# Ajustar el layout para evitar que los títulos se superpongan
plt.tight_layout()
plt.show()
```

### 3. ¿Cuál es la distribución de los clientes por provincia o estado?

- Para provincia, el barplot es el gráfico adecuado para ver la diferencia de clientes entre provincias. Ontario, British Columbia y Quebec son el top 3 con diferencia. 
- Para estado, se ha hecho un pie, pero el Canadá es el único país. 

```python

#Agregación: Contar la frecuencia de clientes por 'Province'
df_province = df_customer_clean['Province'].value_counts().reset_index()
df_province.columns = ['Province', 'Total_Clientes']

plt.figure(figsize=(10, 6))
sns.barplot(
    data=df_province.head(11),
    x='Province',
    y='Total_Clientes',
    palette= 'magma'
)
plt.title('Provincias - Concentración de Clientes')
plt.ylabel('Total de Clientes')
plt.xlabel('Provincia')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Agregación: Contar la frecuencia de clientes por 'Country'
df_country = df_customer_clean['Country'].value_counts()

plt.figure(figsize=(8, 8))

plt.pie(
    df_country, 
    labels=df_country.index, 
    autopct='%1.1f%%', # Muestra el porcentaje con un decimal 
    startangle=90,     # Inicia el primer segmento arriba
    colors=sns.color_palette("Set2") 
)

plt.title('Proporción de Clientes por País')
plt.tight_layout()
plt.show()
```

### 4. ¿Cómo se compara el salario promedio entre los diferentes niveles educativos de los clientes?

Para este caso se han elegido un barplot y un boxplot. 
El barplot permite ver a simple vista las diferencias salariales en función de los estudios, ya que muestra la media. 
Si queremos más detalle, el boxplor nos arroja información más en profundidad. 
    - En el caso de college, solo aparece una línea porque el valor no existía y fue añadido con la media de los valores superior e inferior. La línea muestra esa media, no hay rango de salarios. 

- Clara tendencia: a menores estudios, menos se cobra.
- Mayor diferencia: Quienes tienen un doctorado, ganan el doble que el valor inferior, master. 


```python
plt.figure(figsize=(10, 6))
# 1. Calcular el salario medio (mean) para cada nivel educativo
df_ordenado = df_customer_clean.groupby('Education')['Salary'].mean().reset_index()

# 2. Ordenar los niveles de Education por el salario medio (ascendente)
df_ordenado = df_ordenado.sort_values(by='Salary')

# 3. Extraer el orden para el gráfico (Eje X)
orden_educacion = df_ordenado['Education'].tolist()

#Colores:
colores_para_grafico = palette=sns.color_palette("viridis", n_colors=len(df_customer_clean['Education'].unique()))

# Creamos el Bar Plot
sns.barplot(
    data=df_customer_clean,
    x='Education',
    y='Salary',
    order=orden_educacion,     # Usamos el orden calculado (salario ascendente)
    estimator=np.mean,         
    palette=colores_para_grafico 
)

plt.title('Average Salary by Education Level (Highlighting Bachelor)')
plt.ylabel('Average Salary')
plt.xlabel('Education Level')
plt.xticks(rotation=45, ha='right') # Rotar etiquetas para mejor lectura
plt.tight_layout()
plt.show()
```

```python
#Relación entre Salary y Education
plt.figure(figsize=(12, 6))
sns.boxplot(
    data=df_customer_clean.sort_values('Education'), 
    x='Education', 
    y='Salary',
    palette=sns.color_palette("viridis", n_colors=len(df_customer_clean['Education'].unique())),
    order=orden_educacion
)
plt.title('Distribución Salarial por Nivel Educativo')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.show()
```

### 5. ¿Cuál es la proporción de clientes con diferentes tipos de tarjetas de fidelidad? 

Se quiere contar cuántos clientes hay por cada nivel de "Loyalty Card". 

El gráfico elegido es el Pie Chart, ya que es el mejor para mostrar partes de un todo, porcentajes.
Se podría haber usado también un Bar Plot. 

Casi la mitad de los clientes son "Star". 
Es importante saber el rango de las tarjetas para saber si el programa de fidelización necesita incentivos efectivos para la progresión.





```python

# 1. Agregación: Conteo de clientes por Tarjeta
tarjeta_counts = df_customer_clean['Loyalty Card'].value_counts()

# 2. Creación del Pie Chart
plt.figure(figsize=(8, 8))

plt.pie(
    tarjeta_counts, 
    labels=tarjeta_counts.index, 
    autopct='%1.1f%%', # Muestra el porcentaje (CRUCIAL)
    startangle=90,     
    colors=sns.color_palette("Set2")
)

plt.title('Proporción de Clientes por Tipo de Tarjeta de Fidelidad')
plt.tight_layout()
plt.show()
```

### 6. ¿Cómo se distribuyen los clientes según su estado civil y género?

El gráfico ideal para ver cómo se distribuyen los clientes según estado civil y género es el Count Plot. Permite compararla frecuencia de dos variables categóricas al mismo tiempo. 

Se puede apreciar que la mayoría de los clientes están casados, seguidos de los solteros (la mitad). 

```python

plt.figure(figsize=(10, 6))
sns.countplot(
    data=df_customer_clean,
    x='Marital Status', # Eje X: Estado Civil
    hue='Gender',       # Segmentación: Género
    palette='tab10', 
    order=df_customer_clean['Marital Status'].value_counts().index # Ordenar por frecuencia
)

plt.title('Distribución de Clientes por Estado Civil y Género')
plt.xlabel('Estado Civil')
plt.ylabel('Total de Clientes')
plt.legend(title='Género')
plt.show()
```
